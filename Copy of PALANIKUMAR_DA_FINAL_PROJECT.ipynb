## Result (Handling Missing Values):
**No Missing values found in the DataSet.**

since marks allocated for handling missing values codes are written below:

```python
# Handle missing values (impute or drop)

# filling categorical with mode, numerical with median (or drop if few)
for col in df.columns:
    if df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0],inplace = True)
    else:
        df[col].fillna(df[col].median(),inplace = True)

# Verifying again
df.isnull().sum()
```

```python
# Handle duplicates

# Check duplicates
df.duplicated().sum()

# Drop duplicates
df.drop_duplicates(inplace=True)

# Verifying again
df.duplicated().sum()
```

## Result (Handling Duplicates):
**No Duplicates found in the DataSet.**

since marks allocated for handling Duplicates, codes are written below:

```python
# Treat outliers using IQR (remove extreme values)

def treat_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    # keep only values inside the limits
    df = df[(df[column] >= lower_limit) & (df[column] <= upper_limit)]
    return df
```

## Result (Handling Outliers):
**No outliers found for the numerical columns in the Dataset.**

```python
# Check skewness and apply transformations

# Define numerical columns before checking skewness
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Check skewness
df[num_cols].skew()
```

```python
def log_transform(col):
    return np.log1p(df[col])

def sqrt_transform(col):
    return np.sqrt(df[col])

print("\nData is approximately normally distributed → No transformations required.")
```

## Result(checking Skewness):
**All numeric columns have skewness values between -0.01 and +0.02,
The data is approximately normally distributed.**

No transformations (like log or sqrt) are required for these variables.

Since marks are allocated, codes written below:

```python
# Convert price to float

df['discounted_price'] = (
    df['discounted_price'].astype(str).str.replace("₹", "").str.replace(",", "").astype(float)
)
df['actual_price'] = (
    df['actual_price'].astype(str).str.replace("₹", "").str.replace(",", "").astype(float)
)

# Convert discount percentage to fraction
df['discount_percentage'] = (
    df['discount_percentage'].astype(str).str.replace("%", "").astype(float) / 100
)

# Clean & convert ratings
df['rating'] = df['rating'].astype(str).replace("|", "3.9").astype(float)

# Convert rating_count to numeric
df['rating_count'] = df['rating_count'].astype(str).str.replace(",", "").astype(float)

df.info()
```

```python
# Round ratings to nearest integer for grouping
df['rating_rounded'] = df['rating'].round()

rating_share = df['rating_rounded'].value_counts().sort_index()

# New custom color palette
colors = sns.color_palette("Spectral", n_colors=len(rating_share))

plt.figure(figsize=(8,8))
plt.pie(
    rating_share,
    labels=rating_share.index.astype(str) + " Stars",
    autopct="%1.1f%%",
    startangle=140,
    colors=colors,
    wedgeprops={'edgecolor': 'black'}
)
plt.title("Share of Products by Rating", fontsize=14, fontweight='bold')
plt.show()
```

#Interpretation:


*   Most products fall within the mid-rating range, indicating generally average customer satisfaction.
*   Very high or very low ratings are less common, showing that extreme satisfaction or dissatisfaction is relatively rare.



### Average Rating by Top 10 Product Categories

```python
# Calculate average rating per category
category_ratings = df.groupby('category')['rating'].mean().sort_values(ascending=False)

# Select top 10 categories
top_10_categories = category_ratings.head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_categories.index, y=top_10_categories.values, palette='viridis', hue=top_10_categories.index, legend=False)
plt.title('Average Rating of Top 10 Product Categories', fontsize=16)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Average Rating', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

### Interpretation:
*   This bar chart displays the average customer rating for the top 10 product categories. Each bar represents a category, and its height indicates the average rating received by products within that category.
*   This visualization helps identify which product categories consistently receive high customer satisfaction, potentially guiding inventory and marketing strategies.

### Visualizing Average Discounted Price by Category (from Pivot Table)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the pivot table
pivot_table_category = df.groupby('category')['discounted_price'].mean().to_frame(name='discounted_price')

# Sort the pivot table by discounted_price for better visualization
visual_data = pivot_table_category.sort_values(by='discounted_price', ascending=False).head(15) # Taking top 15 for clarity

plt.figure(figsize=(14, 8))
sns.barplot(
    x=visual_data.index,
    y=visual_data['discounted_price'],
    palette='magma',
    hue=visual_data.index,
    legend=False
)
plt.title('Top Categories by Average Discounted Price', fontsize=16)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Average Discounted Price', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

### Interpretation of Average Discounted Price by Category

**Explain the Chart:**
This bar chart visualizes the average discounted price for the top 15 product categories, derived from our pivot table. Categories are sorted in descending order by their average discounted price, allowing for easy identification of the most expensive and more affordable product groups after discounts.

**What is it saying?**
*   **Pricing Landscape:** The chart clearly shows which categories command higher average discounted prices, giving insight into the general pricing tier of products within those segments.
*   **Value Perception:** While this chart doesn't directly show profitability or sales volume, understanding average discounted prices helps in strategizing promotions and understanding customer expectations for price points in different categories.
